```python
import os
import pandas as pd
from matplotlib import pyplot as plt
import json
import geopandas as gpd
import numpy as np
```

```python
population = pd.read_csv("Parsed/age_group.csv")
population.head()
```

Clean the data

```python
population.index = population.Variable
population.index = population.index.map(str.strip)
population.columns = population.columns.map(str.strip)
population.head()
```

```python
population = population.iloc[:,2:]
total_pop = population.loc[population.index.str.startswith('Total')]
```

```python
with open("total_population_pd_ready.csv", 'r') as f:
    total_pop = json.load(f)
```

```python
total_pop
```

```python
total_pop = total_pop.transpose()
total_age = total_pop.iloc[:,0]
```

```python
total_age_values = total_age.apply(lambda x: ''.join(x.split(',')))
total_age_values = total_age_values.astype('int32')
total_pop.iloc[:,0] = total_age_values
```

```python
total_pop = total_pop.astype('int32')
```

```python
total_pop.to_csv("total_population_pd_ready.csv")
total_pop.to_pickle("total_population_pd_ready.pickle")
```

Now doing population 15 and over

```python
teen_above = population.loc[population['Bottom Age'] >= 15]
```

```python
def get_15_above_population(dataframe, slices=0):
    total_teen = dataframe.iloc[18*slices:18*(slices+1),2:]
    total_teen = total_teen.astype('int32')
    return total_teen.sum(axis=0)
```

```python
total_teen_sum = get_15_above_population(teen_above, slices=0)
total_teen_sum
```

```python
total_teen_male_sum = get_15_above_population(teen_above, slices=1)
```

```python
total_teen_female_sum = get_15_above_population(teen_above, slices=2)
total_teen_female_sum
```

```python
age_15_above = pd.DataFrame({"total_15_above_population": total_teen_sum,
                            "male_15_above_population": total_teen_male_sum,
                            "female_15_above_population": total_teen_female_sum})
```

```python
age_15_above.to_pickle("total_population_age_15_above.pickle")
```

Get 25 to 64 yo sum

```python
adults = population.loc[population['Bottom Age'] >= 15]
adults = adults.loc[adults['Top Age'] <=64]
```

```python
def get_25_64_above_population(dataframe, slices=0):
    total_teen = dataframe.iloc[10*slices:10*(slices+1),2:]
    total_teen = total_teen.astype('int32')
    return total_teen.sum(axis=0)
```

```python
total_adult = get_25_64_above_population(adults, slices=0)
male_adult = get_25_64_above_population(adults, slices=1)
female_adult = get_25_64_above_population(adults, slices=2)
```

```python
adult_age= pd.DataFrame({"total_adult_25_64": total_adult,
                        "male_adult_25_64":male_adult,
                        "female_adult_25_64":female_adult})
adult_age.to_pickle("total_population_adult_25_64.pickle")
```
